```python
#Gernating all pages of  car24 that we want for data :
for i in range(1, 23):
    print('https://www.cars24.com/buy-used-cars-new-delhi/?itm_source=Cars24Website&itm_medium=sticky_header&page={}'.format(i))
```

```python
#here, we make four list for append data in it and getting all the details 
%%time

carname = []
cardetail = []
cartype = []
cardisprice = []


for i in range(1, 23):
    URL = 'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001&page={}'.format(i)
    page = requests.get(URL)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    # carnames
    carnames = soup.find_all('h2', attrs={'class' : '_3FpCg'})
    for item in carnames:
        carname.append(item.text)
        
    # cardetails
    cardetails = soup.find_all('p', attrs={'class' : 'cvakB'})
    for item in cardetails:
        cardetail.append(item.text)
        
    # cartypes
    cartypes = soup.find_all('ul', attrs={'class' : 'bVR0c'})
    for item in cartypes:
        cartype.append(item.text)
        
    # cardisprices
    cardisprices = soup.find_all('div', attrs={'class' : '_7udZZ'})
    for item in cardisprices:
        cardisprice.append(item.text)
```

```python
print(len(carname))
print(len(cardetail))
print(len(cartype))
print(len(cardisprice))
```

```python
df = pd.DataFrame({'carname' : carname, 'cardetail' : cardetail, 'cartype' : cartype, 'cardisprice' : cardisprice})
```

# Cleaning the Data

Spliting the column with the help of regular experssion and condition for make data more use full:-

```python
regex = r'^\d+'

df['Model'] = df['carname'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
regex = r'^\d+,\d+\s\D+'

df['Kilometer'] = df['cartype'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
regex = r'(?:Diesel|Petrol)'

df['fueltype'] = df['cartype'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
regex = r'(?:Automatic|Manual)'
df['carfunc'] = df['cardetail'].apply(lambda x : re.findall(regex, x))
```

```python
df.head()
```

```python
carbrand = []
for i in df.carname:
     carbrand.append(i.split()[1])
```

```python
df['Carbrand'] = carbrand
```

```python
df.head()
```

```python
carmodel = []
for i in df.carname:
     carmodel.append(i.split()[2])
```

```python
df['Carmodel'] = carmodel
```

```python
df.head()
```

```python
df.Model = df.Model.apply(lambda x : ''.join(x))
df.Kilometer = df.Kilometer.apply(lambda x : ''.join(x))
df.fueltype = df.fueltype.apply(lambda x : ''.join(x))
df.carfunc = df.carfunc.apply(lambda x : ''.join(x))
```

```python
df.head()
```

```python
df.info()
```

Drop unnecessary columns:-

```python
df.drop('cartype', axis=1, inplace=True)
df.drop('cardetail', axis=1, inplace=True)
df.drop('carname', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Changing the data type of kilometer , cardisprice and model by removing unwanted things containig the columns

```python
df.Kilometer = df.Kilometer.apply(lambda x : x.replace(',', ''))
df['Kilometer'] = df['Kilometer'].str.replace('km','')
```

```python
df.Model = df.Model.astype(int)
```

```python
df.head()
```

```python
df.cardisprice = df.cardisprice.apply(lambda x : x.replace('₹', '').replace(',', '')).astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
for Model in df.columns:
    if(df[Model].dtype == "int32"):
        df[Model] = df[Model].astype("int64")
```

```python
df.info()
```

```python
Kilometer_use = []
for i in list(df['Kilometer'].values):
    try:
        Kilometer_use.append((float(str(i))))
    except(ValueError):
        Kilometer_use.append(np.nan)
        
df['Kilometer'] = Kilometer_use
```
